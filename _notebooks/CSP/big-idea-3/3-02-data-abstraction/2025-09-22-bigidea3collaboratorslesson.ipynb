  <!-- LESSON 3.2: DATA ABSTRACTION -->
  <section id="l31" class="lesson">
    <header class="hero">
      <div class="hero-inner">
        <h1>3.2 ‚Äî Data Abstraction</h1>
        <p class="sub">Lists & strings, indexes, and how abstraction manages complexity.</p>
      </div>
    </header>

    <aside class="toc" aria-label="Section navigation">
      <nav>
        <ol>
          <li><a href="#l31-obj">Learning Objectives</a></li>
          <li><a href="#l31-ek">Essential Knowledge</a></li>
          <li><a href="#l31-types">New Variable Types</a></li>
          <li><a href="#l31-index">List Index</a></li>
          <li><a href="#l31-da">Data Abstraction ‚Äî Lists</a></li>
          <li><a href="#l31-manage">How Lists Manage Complexity</a></li>
        </ol>
      </nav>
    </aside>

    <main class="content">
      <article id="l31-obj" class="card">
        <h2>Learning Objectives</h2>
        <ul>
          <li>Represent a <strong>list</strong> or <strong>string</strong> using a variable.</li>
          <li>For data abstraction, <strong>develop data abstraction</strong> using lists to store multiple elements.</li>
          <li>Explain how the use of <strong>data abstraction manages complexity</strong> in program code.</li>
        </ul>
      </article>

      <article id="l31-ek" class="card">
        <h2>Essential Knowledge</h2>
        <ul>
          <li>A <strong>list</strong> is an ordered sequence of elements.</li>
          <li>An <strong>element</strong> is an individual value in a list.</li>
          <li>An <strong>index</strong> is a common method for referencing the elements in a list or string using natural numbers.</li>
          <li>A <strong>string</strong> is an ordered sequence of characters.</li>
          <li><strong>Data abstraction</strong> provides a separation between the abstract properties of a data type and the concrete details of its representation.</li>
          <li>Data abstraction can be created using <strong>lists</strong>.</li>
          <li>The exam reference sheet provides notation for lists and describes a list structure whose index values are <strong>1 through the number of elements</strong> (inclusive). For all list operations, if a list index is less than 1 or greater than the length of the list, an <strong>error</strong> is produced and the program will terminate.</li>
        </ul>
      </article>

      <article id="l31-types" class="card">
        <h2>New Types of Variables</h2>
        <div class="grid">
          <div class="snippet">
            <h3>Strings</h3>
            <ul>
              <li>An ordered sequence of characters.</li>
              <li>May contain letters, numbers, and special characters.</li>
              <li>Examples: words, phrases, sentences, ID numbers.</li>
            </ul>
          </div>
          <div class="snippet">
            <h3>Lists</h3>
            <ul>
              <li>An ordered sequence of elements.</li>
              <li>Each element is a variable.</li>
              <li>Examples: playlist of songs, names of students in a class, contacts in your phone.</li>
            </ul>
          </div>
        </div>
      </article>

      <article id="l31-index" class="card">
        <h2>List Index</h2>
        <ul>
          <li>Each element of a list (and each character of a string) is referenced by an <strong>index</strong>.</li>
          <li>For the AP exam, the index starts at <strong>1</strong>.</li>
          <li>Using an index &lt; 1 or &gt; length causes an <strong>error</strong>.</li>
        </ul>
        <pre><code class="language-pseudo">pets ‚Üê ["üê∂","üê±","üêπ","üê¢"]
# Indexes: 1..4 (AP CSP)
# pets[1] = "üê∂"
</code></pre>
      </article>

      <article id="l31-da" class="card">
        <h2>Data Abstraction ‚Äî Lists</h2>
        <ul>
          <li>Lists allow for data abstraction ‚Äî bundle variables together (strings, numbers, characters, etc.).</li>
          <li>Give one name to a set of memory cells.</li>
          <li>You do not need to know in advance how many variables will be needed.</li>
          <li>You do not need to know how the elements are stored together.</li>
        </ul>
      </article>

      <article id="l31-manage" class="card">
        <h2>How Lists Manage Complexity</h2>
        <ul>
          <li><strong>Fewer variables:</strong> one list variable can hold many related items (e.g., all students).</li>
          <li><strong>Dynamic size:</strong> add/remove elements without redefining many variables (e.g., new student transfers).</li>
          <li><strong>Consistent computation:</strong> apply the same computation over all items (e.g., curve all test scores).</li>
        </ul>
      </article>
    </main>
  </section>

    <header class="hero">
      <div class="hero-inner">
        <h1>3.3 ‚Äî Mathematical Expressions</h1>
        <p class="sub">Values, variables, operators, functions ‚Äî expressions evaluate to a single result.</p>
      </div>
    </header>

    <aside class="toc" aria-label="Section navigation">
      <nav>
        <ol>
          <li><a href="#l32-what">What Are Mathematical Expressions?</a></li>
          <li><a href="#l32-ops">Operators</a></li>
          <li><a href="#l32-example">Example: Distance</a></li>
          <li><a href="#l32-key">Key Takeaways</a></li>
          <li><a href="#l32-res">Additional Resources</a></li>
        </ol>
      </nav>
    </aside>

    <main class="content">
      <article id="l32-what" class="card">
        <h2>What Are Mathematical Expressions?</h2>
        <p>In programming, a mathematical expression is a combination of values, variables, operators, and functions that the computer evaluates to produce a single result.</p>
        <pre><code>5 + 3
x * y - z
abs(a - b)</code></pre>
      </article>

      <article id="l32-ops" class="card">
        <h2>Operators</h2>
        <div class="grid">
          <div class="snippet">
            <pre><code>+   addition
-   subtraction
*   multiplication
/   division
**  exponents
%   modulo (remainder)
</code></pre>
            <p><em>Note:</em> In Python, MOD is represented by <code>%</code>.</p>
          </div>
        </div>
      </article>

      <article id="l32-example" class="card">
        <h2>Example: Calculating Distance</h2>
        <p>Distance between two points (x‚ÇÅ, y‚ÇÅ) and (x‚ÇÇ, y‚ÇÇ):</p>
        <pre><code class="language-tex">distance = sqrt((x2 - x1)^2 + (y2 - y1)^2)</code></pre>
        <pre><code class="language-pseudo">diffX ‚Üê x2 - x1
diffY ‚Üê y2 - y1
distance ‚Üê sqrt(diffX * diffX + diffY * diffY)</code></pre>
      </article>

      <article id="l32-key" class="card">
        <h2>Key Takeaways</h2>
        <ul>
          <li>Expressions evaluate to a single value.</li>
          <li>Operators perform operations on values or variables.</li>
          <li>Functions can be used to perform more complex calculations.</li>
          <li>Understanding expressions is crucial for developing algorithms and controlling program behavior.</li>
        </ul>
      </article>

      <article id="l32-res" class="card">
        <h2>Additional Resources</h2>
        <ul>
          <li>Mathematical Expressions Study Guide (Fiveable)</li>

        </ul>
        <pre><code>https://fiveable.me/ap-comp-sci-p/unit-3/mathematical-expressions/study-guide/00lGBdF7QyY5hmd1rubD
https://apcentral.collegeboard.org/media/pdf/ap-computer-science-principles-course-and-exam-description.pdf</code></pre>
      </article>
    </main>
  </section>

  <!-- LESSON 3.5: BOOLEAN EXPRESSIONS -->
  <section id="l35" class="lesson">
    <header class="hero">
      <div class="hero-inner">
        <h1>3.5 ‚Äî Boolean Expressions & Logic</h1>
        <p class="sub">Booleans, relational operators, modulus, and logical operators (NOT, AND, OR).</p>
      </div>
    </header>

    <aside class="toc" aria-label="Section navigation">
      <nav>
        <ol>
          <li><a href="#l35-bool">Boolean Values</a></li>
          <li><a href="#l35-rel">Relational Operators</a></li>
          <li><a href="#l35-mod">Modulus Operator (%)</a></li>
          <li><a href="#l35-log">Logical Operators</a></li>
        </ol>
      </nav>
    </aside>

    <main class="content">
      <article id="l35-bool" class="card">
        <h2>Boolean Values</h2>
        <p>Booleans represent two states: <strong>True</strong> or <strong>False</strong>. They often come from evaluating expressions (e.g., <code>5 &lt; 10 ‚Üí True</code>) and are used in if-statements, loops, and decision making.</p>
        <pre><code class="language-python">myHairIsBlack = True
iHaveADog = False</code></pre>
        <p><strong>Key takeaway:</strong> Booleans let a program decide what to do based on conditions.</p>
      </article>

      <article id="l35-rel" class="card">
        <h2>Relational Operators</h2>
        <div class="grid">
          <div class="snippet">
            <pre><code>==   Equal           (5 == 5)  ‚Üí True
!=   Not equal       (4 != 5)  ‚Üí True
&gt;    Greater than    (7 &gt; 3)   ‚Üí True
&lt;    Less than       (2 &lt; 8)   ‚Üí True
&gt;=   Greater or equal (6 &gt;= 6) ‚Üí True
&lt;=   Less or equal    (9 &lt;= 4) ‚Üí False</code></pre>
          </div>
        </div>
        <p><em>College Board expectations:</em> recognize/apply relational operators, predict Boolean results, and use them within if-statements and algorithms.</p>
      </article>

      <article id="l35-mod" class="card">
        <h2>Modulus Operator (%)</h2>
        <p>Finds the remainder after division ‚Äî great for even/odd tests and cycles.</p>
        <pre><code class="language-python">print(10 % 2)  # 0 (even)
print(7 % 2)   # 1 (odd)</code></pre>
        <ul>
          <li><code>num % 2 == 0</code> ‚Üí even</li>
          <li><code>num % 2 == 1</code> ‚Üí odd</li>
          <li><code>day % 7</code> ‚Üí day of week in a cycle</li>
        </ul>
        <p><em>College Board expectations:</em> determine remainders, apply modulus in conditions, write Boolean expressions with <code>%</code> to detect patterns.</p>
      </article>

      <article id="l35-log" class="card">
        <h2>Logical Operators</h2>
        <div class="grid">
          <div class="snippet">
            <h3>NOT (negation)</h3>
            <pre><code class="language-python">isCloudy = True
print(not isCloudy)  # False</code></pre>
          </div>
          <div class="snippet">
            <h3>AND (conjunction)</h3>
            <pre><code class="language-python">didHomework = True
scored80 = True
canGoOut = didHomework and scored80  # True</code></pre>
          </div>
          <div class="snippet">
            <h3>OR (disjunction)</h3>
            <pre><code class="language-python">isWeekend = True
hasHoliday = False
dayOff = isWeekend or hasHoliday  # True</code></pre>
          </div>
        </div>
        <p><em>College Board expectations:</em> apply NOT/AND/OR, predict outcomes of compound expressions, and recognize how combinations support decisions.</p>
      </article>
    </main>
  </section>

  <section id="l317" class="lesson">
    <header class="hero">
      <div class="hero-inner">
        <h1>3.17 ‚Äî Algorithms & Efficiency</h1>
        <p class="sub">Problems vs. instances, Big-O growth, reasonable vs. unreasonable time, heuristics.</p>
      </div>
    </header>

    <aside class="toc" aria-label="Section navigation">
      <nav>
        <ol>
          <li><a href="#l317-bigo">Big-O Notation Basics</a></li>
          <li><a href="#l317-problem">What is a Problem?</a></li>
          <li><a href="#l317-instance">What is an Instance?</a></li>
          <li><a href="#l317-types">Types of Problems</a></li>
          <li><a href="#l317-eff">Measuring Efficiency</a></li>
          <li><a href="#l317-reason">Reasonable vs. Unreasonable Time</a></li>
          <li><a href="#l317-growth">Example: Algorithm Growth</a></li>
          <li><a href="#l317-heur">Heuristic Approaches</a></li>
          <li><a href="#l317-tsp">Traveling Salesperson Problem (TSP)</a></li>
          <li><a href="#l317-hs">Example: High School Scheduling</a></li>
          <li><a href="#l317-qc">Quick Check</a></li>
          <li><a href="#l317-gloss">Quick Glossary</a></li>
        </ol>
      </nav>
    </aside>

    <main class="content">
      <article id="l317-bigo" class="card">
        <h2>Big-O Notation Basics</h2>
        <pre><code>O(1)   Constant      ‚Üí same time regardless of input size
O(n)   Linear        ‚Üí grows proportionally with input size
O(n¬≤)  Quadratic     ‚Üí grows with square of input size
O(2‚Åø)  Exponential   ‚Üí explodes rapidly as input increases</code></pre>
      </article>

      <article id="l317-problem" class="card">
        <h2>What is a Problem?</h2>
        <p>A problem is a general task that we want an algorithm to solve.</p>
        <p><em>Example:</em> Sorting is a problem (putting numbers in order).</p>
      </article>

      <article id="l317-instance" class="card">
        <h2>What is an Instance?</h2>
        <p>An instance is a specific case of a problem with actual input values.</p>
        <p><em>Example:</em> Sorting (2, 3, 1, 7) is one instance of the sorting problem.</p>
      </article>

      <article id="l317-types" class="card">
        <h2>Types of Problems</h2>
        <ul>
          <li><strong>Decision Problem:</strong> Yes/No answer. Example: ‚ÄúIs there a path from A to B?‚Äù</li>
          <li><strong>Optimization Problem:</strong> Find the best solution. Example: shortest path from A to B.</li>
        </ul>
      </article>

      <article id="l317-eff" class="card">
        <h2>Measuring Efficiency</h2>
        <p>Efficiency describes how fast an algorithm‚Äôs cost grows as input size increases (count operations, not wall-clock time).</p>
      </article>

      <article id="l317-reason" class="card">
        <h2>Reasonable vs. Unreasonable Time</h2>
        <ul>
          <li><strong>Reasonable:</strong> O(1), O(log n), O(n), O(n log n), O(n¬≤*)</li>
          <li><strong>Unreasonable:</strong> O(2‚Åø), O(n!).</li>
        </ul>
        <p>Quadratic (O(n¬≤)) can be okay for small/medium <em>n</em>.</p>
      </article>

      <article id="l317-growth" class="card">
        <h2>Example: Algorithm Growth</h2>
        <pre><code>Input n | Alg1 (5n) | Alg2 (3‚Åø) | Alg3 (4n¬≤) | Alg4 (n!)
   1    |     5     |     3     |     4      |   1
   2    |    10     |     9     |    16      |   2
   3    |    15     |    27     |    36      |   6
   4    |    20     |    81     |    64      |  24
   5    |    25     |   243     |   100      | 120</code></pre>
      </article>

      <article id="l317-heur" class="card">
        <h2>Heuristic Approaches</h2>
        <p>A heuristic is a smart shortcut that finds a good-enough solution efficiently ‚Äî not always optimal, but practical when exact methods are too slow.</p>
      </article>

      <article id="l317-tsp" class="card">
        <h2>Traveling Salesperson Problem (TSP)</h2>
        <ul>
          <li>Visit all cities once, return to start ‚Üí minimize distance.</li>
          <li>Routes grow factorially:
            <ul>
              <li>4 cities ‚Üí 6 routes</li>
              <li>10 cities ‚Üí 362,880 routes</li>
              <li>20 cities ‚Üí ~10¬π‚Å∑ routes</li>
            </ul>
          </li>
          <li>Heuristics help when exact solutions are impossible for large <em>n</em> (e.g., Nearest Neighbor).</li>
        </ul>
      </article>

      <article id="l317-hs" class="card">
        <h2>Example: High School Scheduling</h2>
        <p><strong>Best use of a heuristic:</strong> creating the master schedule (complex & constrained).</p>
        <p><em>Not heuristic-needed for:</em> searching prices, calculating GPA, generating passwords.</p>
      </article>

      <article id="l317-qc" class="card">
        <h2>Quick Check</h2>
        <p>In which situation is a heuristic most useful?</p>
        <ul>
          <li>A. Searching a list of prices for the smallest value</li>
          <li><strong>‚úÖ B. Creating the master schedule for a high school</strong></li>
          <li>C. Calculating GPA for all students</li>
          <li>D. Generating passwords</li>
        </ul>
      </article>

      <article id="l317-gloss" class="card">
        <h2>Quick Glossary</h2>
        <ul>
          <li><strong>Problem</strong> ‚Äî a general task an algorithm aims to solve.</li>
          <li><strong>Instance</strong> ‚Äî a specific case of a problem.</li>
          <li><strong>Decision Problem</strong> ‚Äî Yes/No answer.</li>
          <li><strong>Optimization Problem</strong> ‚Äî best solution.</li>
          <li><strong>Efficiency / Big-O</strong> ‚Äî how cost grows with input size.</li>
          <li><strong>Heuristic</strong> ‚Äî shortcut method, not guaranteed optimal.</li>
          <li><strong>TSP</strong> ‚Äî classic optimization problem (visit all cities, shortest route).</li>
        </ul>
      </article>
    </main>
  </section>

<script>
(function(){
  const root = document.getElementById('csp-lessons');
  if(!root) return;

  // Parallax shimmer follows cursor (global)
  root.addEventListener('mousemove', (e)=>{
    const r = root.getBoundingClientRect();
    const mx = ((e.clientX - r.left)/r.width * 100).toFixed(2) + '%';
    const my = ((e.clientY - r.top)/r.height * 100).toFixed(2) + '%';
    root.style.setProperty('--mx', mx);
    root.style.setProperty('--my', my);
  });

  // Reveal cards on scroll
  const cards = root.querySelectorAll('.card');
  const io = new IntersectionObserver((entries)=>{
    entries.forEach(e=>{
      if(e.isIntersecting){ e.target.classList.add('reveal','inview'); }
      else { e.target.classList.remove('inview'); }
    });
  }, { threshold: 0.35 });
  cards.forEach(c=>io.observe(c));

  // Global scroll progress + accent hue shift
  const progressBar = root.querySelector('.ui-progress-bar');
  const onScroll = ()=>{
    const max = document.documentElement.scrollHeight - window.innerHeight;
    const pct = Math.max(0, Math.min(1, window.scrollY / max));
    progressBar.style.width = (pct*100).toFixed(1)+'%';
    const hue = 205 + pct*120;
    root.style.setProperty('--accentH', hue.toFixed(1));
  };
  window.addEventListener('scroll', onScroll, { passive:true });
  onScroll();

  // Per-lesson sticky TOC highlighting
  const tocLinks = Array.from(root.querySelectorAll('.toc a'));
  const sectionTargets = tocLinks.map(a => document.querySelector(a.getAttribute('href')));
  const tocIO = new IntersectionObserver((entries)=>{
    entries.forEach(ent=>{
      if(ent.isIntersecting){
        const id = '#' + ent.target.id;
        tocLinks.forEach(l => l.classList.toggle('active', l.getAttribute('href')===id));
      }
    });
  }, { rootMargin: '-40% 0px -50% 0px', threshold: 0.01 });
  sectionTargets.forEach(s => s && tocIO.observe(s));

  // Add copy buttons to all <pre>
  root.querySelectorAll('pre').forEach(pre=>{
    const btn = document.createElement('button');
    btn.className = 'copy-btn';
    btn.type = 'button';
    btn.textContent = 'Copy';
    btn.addEventListener('click', async ()=>{
      const code = pre.innerText.replace(/^\s+|\s+$/g,'');
      try { await navigator.clipboard.writeText(code); btn.textContent = 'Copied!'; }
      catch { btn.textContent = 'Error'; }
      setTimeout(()=> btn.textContent = 'Copy', 1100);
    });
    pre.appendChild(btn);
  });

  // Back-to-top visibility + smooth scroll
  const topBtn = root.querySelector('.to-top');
  const onScrollTop = ()=>{
    const show = window.scrollY > 600; topBtn.classList.toggle('show', show);
  };
  window.addEventListener('scroll', onScrollTop, { passive:true });
  onScrollTop();
  topBtn.addEventListener('click', ()=> window.scrollTo({ top:0, behavior:'smooth' }));
})();
</script>
