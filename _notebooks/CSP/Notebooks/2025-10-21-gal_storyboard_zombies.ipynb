## Sports World: The Algorithm Arena

Exploring performance, data, and design through sports and code.

### Our Route

San Diego → Los Angeles → San Francisco → Seattle

---

### Stop 1: San Diego — “The Data Kickoff”

**Stadiums:** Petco Park (Baseball – Padres), Snapdragon Stadium (Soccer – SDFC)

**Concept:** Collecting and Interpreting Sports Data

**Activity:**
Gather real-time or sample data from both stadiums—attendance, scores, ticket prices, or fan statistics—using APIs or datasets.

**Key Learning:**
Just like scouts tracking player stats, programmers begin by collecting and cleaning raw data before generating insights or predictions.

---

### Stop 2: Los Angeles — “Designing the Playbook”

**Stadiums:** Dodger Stadium (Baseball), SoFi Stadium (Football – Chargers)

**Concept:** Aesthetic UI and Team Branding

**Activity:**
Design dashboards using team colors, logos, and typography.
Incorporate charts, graphs, and interactive components to make the data engaging and easy to interpret.

**Key Learning:**
Design influences understanding—well-crafted visuals help users connect to the data story and retain insights more effectively.

---

### Stop 3: San Francisco — “Charting the Game”

**Stadiums:** Chase Center (Basketball), Oracle Park (Baseball)

**Concept:** Data Visualization and Comparison

**Activity:**
Use programming structures such as variables, conditionals, and functions to manage and compare team or stadium data.
Example functions: `getStadiumInfo("Chase Center")`, `filterBySport("Basketball")`.

**Key Learning:**
Just as reviewing game footage reveals patterns and tactics, visualization highlights performance trends and relationships in data.

---

### Stop 4: Seattle — “The Final Score and Reflection”

**Stadiums:** Lumen Field (Football – Seahawks), T-Mobile Park (Baseball – Mariners)

**Concept:** Iteration and Continuous Improvement

**Activity:**
Compare results from all cities and sports to identify overall performance trends.
Conclude with a dashboard or interactive quiz that demonstrates your findings.

**Key Learning:**
Each dataset, like each season, teaches new lessons. Iteration—refining logic, visuals, and conclusions—mirrors how teams adapt and improve over time.

---
