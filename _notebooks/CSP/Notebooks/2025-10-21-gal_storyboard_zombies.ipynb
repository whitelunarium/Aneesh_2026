## ⚽ Sports World: The Algorithm Arena

**Exploring performance, data, and design through sports and code.**

🗺️ **Our Route**
San Diego → Los Angeles → San Francisco → Santa Clara

---

### 🌴 Stop 1: San Diego — “The Data Kickoff”

🏟️ **Petco Park (Baseball – Padres)**
🏟️ **Snapdragon Stadium (Soccer – SDFC)**

**Concept:** Collecting & Interpreting Sports Data

**Activity:** Gather real-time stats from both stadiums using APIs or provided datasets.
Example: Attendance, team performance, or ticket prices.

💡 **Key Learning:**
Like scouts gathering player stats, programmers collect and clean data before they can analyze or visualize it.

---

### 🎬 Stop 2: Los Angeles — “Designing the Playbook”

🏟️ **Dodger Stadium (Baseball)**
🏟️ **SoFi Stadium (Football – Chargers)**

**Concept:** Aesthetic UI & Team Branding

**Activity:** Create dashboards that use **team colors, logos, and typography** for each team’s data display.
Add **charts, graphs**, and **interactive buttons** for user control.

💡 **Key Learning:**
Design affects engagement — a well-designed interface helps users connect emotionally to the data they see.

---

### 🌉 Stop 3: San Francisco — “Charting the Game”

🏟️ **Chase Center (Basketball)**
🏟️ **Oracle Park (Baseball)**

**Concept:** Data Visualization & Comparison

**Activity:**
Use **variables, conditionals, and functions** to build comparisons between teams or stadiums.
Example: `filterBySport("Basketball")` or `getStadiumInfo("Chase Center")`.

💡 **Key Learning:**
Like analyzing plays on a screen, visualization helps spot patterns and tell the story behind the numbers.

---

### 🏈 Stop 4: Santa Clara — “The Final Score”

🏟️ **Levi’s Stadium (Football – 49ers)**

**Concept:** Data-Driven Decisions & Iteration

**Activity:**
Compare multiple stadiums and sports — use your data to make **predictions** or design a **mini-quiz** for users.

💡 **Key Learning:**
Every analysis leads to iteration — refining your logic, improving your visuals, and learning from the data’s story.

---
