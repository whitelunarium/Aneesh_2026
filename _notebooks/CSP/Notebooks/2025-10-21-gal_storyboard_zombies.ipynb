## Storyboard Summary!

Exploring how APIs bring sports statistics and stadium data to life through performance, logic, and visualization.

### Our Route

San Diego → Los Angeles → San Francisco → Seattle

--- 

### Stop 1: San Diego — “Connecting to the Data Field”

**Focus:** Petco Park (Baseball – Padres) and Snapdragon Stadium (Soccer – SDFC)

**Coding Concept:** API Handling — GET and POST Requests

**Activity:**
Use APIs to collect **team and stadium data**, such as scores, ticket prices, fan attendance, or upcoming schedules.
Display the retrieved data in a simple interface or dashboard.

**Key Learning:**
APIs act like scouts—fetching live data from multiple sources and delivering it to your program in real time.

---

### Stop 2: Los Angeles — “Filtering the Playbook”

**Focus:** Dodger Stadium (Baseball – Dodgers) and SoFi Stadium (Football – Chargers)

**Coding Concept:** Sorting and Filtering API Data

**Activity:**
Query the same API endpoints but apply filters—for example, retrieve only home game stats, specific dates, or certain player records.
Visualize how filtering reveals deeper performance insights.

**Key Learning:**
Just as analysts focus on key plays or metrics, coders use filtering logic to extract the most relevant information from an API.

---

### Stop 3: San Francisco — “Analyzing Performance Data”

**Focus:** Chase Center (Basketball – Warriors) and Oracle Park (Baseball – Giants)

**Coding Concept:** Combining and Comparing API Responses

**Activity:**
Use multiple API calls to pull **both team and stadium stats**, then merge or compare the data sets.
Create charts that show relationships—like average attendance vs. win rate or home vs. away performance.

**Key Learning:**
By combining datasets, programmers uncover patterns—just as coaches use statistics to analyze performance trends.

---

### Stop 4: Seattle — “Updating the Leaderboard”

**Focus:** Lumen Field (Football – Seahawks) and T-Mobile Park (Baseball – Mariners)

**Coding Concept:** Fetching and Updating Data with API Operations

**Activity:**
Build logic for updating or posting new information, such as editing records, adding new match results, or recalculating averages.
Reflect updates in a dynamic leaderboard or summary dashboard.

**Key Learning:**
APIs don’t just read data—they let you write and update it, keeping your application as current as the live game feed.

---
