# 📗 Lesson: Strings in JavaScript
---
## 1. Creating Strings
```javascript
let name = "Alice";
let greeting = 'Hello';
let multiline = `This is
a template string`;
```
- Supports single `'`, double `"`, and backticks `` ` `` for template literals.

````javascript
// Example: Creating strings
let name = "Alice";
let greeting = 'Hello';
console.log(name, greeting);
````

## 2. Combining Strings
```javascript
let first = "Hello";
let second = "World";
console.log(first + " " + second); 
// Output: Hello World
```

````javascript
let first = "Hello";
let second = "World";
console.log(first + " " + second);
````

## 3. String Functions & Properties
```javascript
let text = "javascript is fun";
console.log(text.length);         // 17
console.log(text.toUpperCase());  // JAVASCRIPT IS FUN
console.log(text.replace("fun", "awesome")); // javascript is awesome
```

````javascript
let text = "javascript is fun";
console.log(text.length);
console.log(text.toUpperCase());
console.log(text.replace("fun", "awesome"));
````
